```python
client_df =df_delivered.groupby('Телефон_new').agg(
    delivered_orders=('НомерЗаказаНаСайте', 'nunique'),
    items=('КоличествоПроданоКлиенту', 'sum'),
    revenue=('СуммаСтроки', 'sum'),
    margin=('Маржа', 'sum'),
    first_purchase=('Дата', 'min'),                                                            
    last_purchase=('Дата', 'max'),                                     
#     delivery_methods=('МетодДоставки', lambda x: x.mode()[0] if not x.mode().empty else 'Нет'),             
#     payment_methods=('ФормаОплаты', lambda x: x.mode()[0] if not x.mode().empty else 'Нет') 
    
).reset_index()
client_df['avg_order_value'] = client_df['revenue'] / client_df['delivered_orders']
client_df = client_df.sort_values(by='revenue', ascending=False)
```

```python
# Самое частое колво товаров на складе
order_sizes = (
    df_delivered.groupby(['Телефон_new', 'НомерЗаказаНаСайте'])['КоличествоПроданоКлиенту'].sum()
    .reset_index(name='order_size')
)

most_freq_order_size = (
    order_sizes.groupby('Телефон_new')['order_size']
    .agg(lambda x: x.mode()[0] if not x.mode().empty else None)
    .reset_index(name='most_freq_order_size')
)
```

```python
# recency = today_date – last_purchase
# days_since_first_purchase = дни с момента первой покупки ()
today = df_delivered['Дата'].max()

recency_df = (
    df_delivered.groupby('Телефон_new')['Дата']
    .agg(
        last_purchase_date='max',
        first_purchase_date='min'
    )
    .reset_index()
)

# days since last purchase
recency_df['recency_days'] = (today - recency_df['last_purchase_date']).dt.days

# days since first purchase
recency_df['days_since_first_purchase'] = (today - recency_df['first_purchase_date']).dt.days
```

```python
# IQR outliers
key_cols = [
    'revenue',
    'margin', 
    'avg_order_value', 
]

Q1 = client_df[key_cols].quantile(0.1)
Q3 = client_df[key_cols].quantile(0.9)
IQR = Q3 - Q1

outlier_mask = (
    (client_df[key_cols] < (Q1 - 3 * IQR)) |
    (client_df[key_cols] > (Q3 + 3 * IQR))
)

outlier_rows = outlier_mask.sum(axis=1) >= 2
client_df = client_df.loc[~outlier_rows]
```

```python
# 3. Случайные данные для Gap
random_data = np.random.random_sample(df_scaled.shape)

from tqdm import tqdm

k_range = range(5, 22)
sse, silhouette_scores, dbi_scores, ch_scores, gap_values = [], [], [], [], []

for k in tqdm(k_range, desc="Computing cluster metrics"):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    labels = kmeans.labels_
    
    # SSE (Elbow)
    sse.append(kmeans.inertia_)
    
    # Silhouette
    silhouette_scores.append(silhouette_score(df_scaled, labels))
    
    # Davies–Bouldin
    dbi_scores.append(davies_bouldin_score(df_scaled, labels))
    
    # Calinski–Harabasz
    ch_scores.append(calinski_harabasz_score(df_scaled, labels))
    
    # Gap
    random_kmeans = KMeans(n_clusters=k, random_state=42, n_init=10).fit(random_data)
    gap_values.append(
        np.log(np.mean(cdist(random_data, random_kmeans.cluster_centers_, 'euclidean'))) -
        np.log(np.mean(cdist(df_scaled, kmeans.cluster_centers_, 'euclidean')))
    )
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

# df_scaled должен быть готов (numpy array или DataFrame только с числами)
X = np.asarray(df_scaled)

k_range = range(5, 22)
B = 10  # число reference-выборок для Gap (10–20 обычно достаточно)

rng = np.random.default_rng(42)

# bounds для равномерного сэмплинга в "bounding box" данных
mins = X.min(axis=0)
maxs = X.max(axis=0)

sse, sil, dbi, ch, gap = [], [], [], [], []

for k in tqdm(list(k_range), desc="Computing metrics + Gap"):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_k = km.fit_predict(X)

    # базовые метрики
    sse.append(km.inertia_)
    sil.append(silhouette_score(X, labels_k))
    dbi.append(davies_bouldin_score(X, labels_k))
    ch.append(calinski_harabasz_score(X, labels_k))

    # Gap statistic (Tibshirani et al.)
    # Wk = within-cluster dispersion; в sklearn inertia_ = sum of squared distances to centers
    log_wk = np.log(km.inertia_)

    ref_log_wk = []
    for _ in range(B):
        X_ref = rng.uniform(mins, maxs, size=X.shape)
        km_ref = KMeans(n_clusters=k, random_state=42, n_init=10).fit(X_ref)
        ref_log_wk.append(np.log(km_ref.inertia_))

    gap.append(np.mean(ref_log_wk) - log_wk)

# --- один общий график (нормализованные кривые) ---

# делаем "выше = лучше" для всех кривых:
# SSE и DBI лучше ниже, поэтому инвертируем
# sse_inv = -np.array(sse, dtype=float)
# dbi_inv = -np.array(dbi, dtype=float)

metrics = np.vstack([
    sse_inv,
    np.array(sil, dtype=float),
    dbi_inv,
    np.array(ch, dtype=float),
    np.array(gap, dtype=float),
])

names = ["SSE (inverted)", "Silhouette", "DBI (inverted)", "CH", "Gap"]

# нормализация 0–1 по каждой метрике
mn = metrics.min(axis=1, keepdims=True)
mx = metrics.max(axis=1, keepdims=True)
den = np.where((mx - mn) == 0, 1, (mx - mn))
metrics_norm = (metrics - mn) / den

plt.figure(figsize=(12, 6))
for i, name in enumerate(names):
    plt.plot(list(k_range), metrics_norm[i], marker='o', linestyle='--', label=name)

plt.title("Cluster Metrics (normalized, higher = better)")
plt.xlabel("k")
plt.ylabel("Normalized value (0–1)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from tqdm import tqdm

k_range = range(5, 35)

bic, aic, ll = [], [], []

for k in tqdm(k_range, desc="EM model selection"):
    gmm = GaussianMixture(
        n_components=k,
        covariance_type="full",
        random_state=42,
        n_init=5
    )
    gmm.fit(df_scaled)

    bic.append(gmm.bic(df_scaled))
    aic.append(gmm.aic(df_scaled))
    ll.append(gmm.score(df_scaled))  # avg log-likelihood
```

```python
from scipy.cluster.hierarchy import linkage
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

X = df_scaled.to_numpy()
n, d = X.shape
print("Shape:", X.shape)

# sample for Ward 
rng = np.random.default_rng(42)
sample_size = 10000  # 5k-15k

sample_idx = rng.choice(n, size=sample_size, replace=False)
X_s = X[sample_idx]

print("Sample shape:", X_s.shape)
```

```python
# Ward linkage on sample
Z = linkage(X_s, method="ward")

import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(14, 6))
dendrogram(
    Z,
    truncate_mode="lastp",   # показываем последние p "суперкластеров"
    p=40,                    # 30–60 обычно норм
    show_contracted=True,    # сворачивает поддеревья
    leaf_rotation=0,
    leaf_font_size=10
)
plt.title("Ward dendrogram (truncated)")
plt.xlabel("Collapsed clusters")
plt.ylabel("Merge height (Ward)")
plt.tight_layout()
plt.show()
```

```python
heights = Z[:, 2]  # высота каждого слияния

gaps = np.diff(heights)  # разницы между соседними слияниями

# покажем хвост (там обычно самые важные скачки)
plt.figure(figsize=(12, 4))
plt.plot(heights[-200:])  # последние 200 слияний
plt.title("Last merges heights (Ward)")
plt.ylabel("Merge height")
plt.xlabel("Merge step (tail)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 4))
plt.plot(gaps[-200:])
plt.title("Gaps between last merges (Ward)")
plt.ylabel("Gap")
plt.xlabel("Merge step (tail)")
plt.tight_layout()
plt.show()
```

```python
t = 115

plt.figure(figsize=(14, 6))
dendrogram(Z, truncate_mode="lastp", p=40, show_contracted=True)
plt.axhline(y=t, linewidth=2)  # линия разреза
plt.title(f"Ward dendrogram cut at t={t}")
plt.ylabel("Merge height")
plt.tight_layout()
plt.show()
```

```python
def distribution_report(s: pd.Series, name: str):
    s_clean = s.dropna()
    n = len(s_clean)
    zeros = (s_clean == 0).sum()
    q = s_clean.quantile([0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99])
    print(f"\n{name}:")
    print(f"  n={n}")
    print(f"  zeros={zeros} ({zeros/n:.1%})")
    print(f"  min={s_clean.min():.4g}, max={s_clean.max():.4g}")
    print("  quantiles:")
    for k, v in q.items():
        print(f"    {k:>4.2f}: {v:.4g}")
    # скошенность (на больших данных полезно)
    try:
        print(f"  skew={s_clean.skew():.3f}")
    except Exception:
        pass

distribution_report(rfm["R_recency_days"], "R (recency_days) — меньше = лучше")
distribution_report(rfm["F_delivered_orders"], "F (delivered_orders) — больше = лучше")
distribution_report(rfm["M_revenue"], "M (revenue) — больше = лучше")
```

```python
# R-score (1..5): меньше recency => лучше, поэтому сначала qcut, потом инверсия
rfm["R_score"] = pd.qcut(rfm["R_recency_days"], 5, labels=[1,2,3,4,5]).astype(int)
rfm["R_score"] = 6 - rfm["R_score"]   # инверсия: 5 = самый свежий

# M-score (1..5): больше revenue => лучше
rfm["M_score"] = pd.qcut(rfm["M_revenue"], 5, labels=[1,2,3,4,5]).astype(int)

# F-score: по правилам под эти данные
# 1 заказ -> 1, 2 заказа -> 3, 3+ -> 5
rfm["delivered_orders"] = rfm["F_delivered_orders"].astype(int)
rfm["F_score"] = np.where(rfm["F_delivered_orders"] == 1, 1,
                   np.where(rfm["F_delivered_orders"] == 2, 3, 5))
```

```python
# RFM-код
rfm["RFM_code"] = (
    rfm["R_score"].astype(str) +
    rfm["F_score"].astype(str) +
    rfm["M_score"].astype(str)
)

print("R_score counts:\n", rfm["R_score"].value_counts().sort_index(), "\n")
print("F_score counts:\n", rfm["F_score"].value_counts().sort_index(), "\n")
print("M_score counts:\n", rfm["M_score"].value_counts().sort_index(), "\n")
```

```python
rfm["segment"] = "Other"

# Champions: свежие + повторные/лояльные + дорогие
rfm.loc[(rfm["R_score"] >= 4) & (rfm["F_score"] >= 3) & (rfm["M_score"] >= 4), "segment"] = "Champions"

# Loyal: много заказов (F_score=5 это 3+ заказа) и не слишком старые
rfm.loc[(rfm["R_score"] >= 3) & (rfm["F_score"] == 5), "segment"] = "Loyal"

# Big Spenders: топ по деньгам, даже если не супер часто (только если ещё Other)
rfm.loc[(rfm["M_score"] == 5) & (rfm["R_score"] >= 3) & (rfm["segment"] == "Other"), "segment"] = "Big Spenders"

# New: самые свежие и 1 заказ
rfm.loc[(rfm["R_score"] == 5) & (rfm["F_score"] == 1), "segment"] = "New"

# Promising: свежие, но пока не лояльные, деньги средние
rfm.loc[(rfm["R_score"] >= 4) & (rfm["F_score"] <= 3) & (rfm["M_score"].between(2, 4)) & (rfm["segment"] == "Other"),
        "segment"] = "Promising"

# At Risk: давно не покупали, но деньги высокие
rfm.loc[(rfm["R_score"] <= 2) & (rfm["M_score"] >= 4), "segment"] = "At Risk"

# About to Sleep: давно и деньги средние/низкие (только если Other)
rfm.loc[(rfm["R_score"] <= 2) & (rfm["M_score"] <= 3) & (rfm["segment"] == "Other"), "segment"] = "About to Sleep"

# Lost: совсем старые и одноразовые
rfm.loc[(rfm["R_score"] == 1) & (rfm["F_score"] == 1), "segment"] = "Lost"


# Сколько людей в сегментах
print("Segment counts:\n", rfm["segment"].value_counts(), "\n")
```

```python
# Профиль сегментов (сколько клиентов и денег)
profile = (
    rfm.groupby("segment")
       .agg(
           customers=("segment", "size"),
           revenue_sum=("M_revenue", "sum"),
           revenue_mean=("M_revenue", "mean"),
           orders_mean=("F_delivered_orders", "mean"),
           recency_mean=("R_recency_days", "mean"),
           recency_median=("R_recency_days", "median"),
       )
       .sort_values("revenue_sum", ascending=False)
)

print("Segment profile (sorted by total revenue):")
display(profile)
```

```python
# Доли клиентов и выручки
profile2 = profile.copy()
profile2["customers_share_%"] = 100 * profile2["customers"] / profile2["customers"].sum()
profile2["revenue_share_%"] = 100 * profile2["revenue_sum"] / profile2["revenue_sum"].sum()

display(profile2.sort_values("revenue_share_%", ascending=False))
```
